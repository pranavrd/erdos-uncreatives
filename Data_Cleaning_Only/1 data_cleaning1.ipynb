In this notebook we do some basic data cleaning. Our data arrived in batches so we need to concatenate the data frames together into one data set. 
We then dropped a bunch of columns that did not contain anything of interest to us or contained garbage information. We also dropped any videos that may have been accidentally scraped twice. 

```python
import os
import glob
import pandas as pd
```

```python
#Use glob to get CSV file paths
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
```

```python
#read and concatenate all CSV files
df_list = [pd.read_csv(file) for file in csv_files]
concatenated_df = pd.concat(df_list, ignore_index=True)
```

```python

# Drop rows with missing and duplicated data in column: 'id' this column contains the unique identifier for each video
#if data is missing that means that the information does not pertain to a video, mostly is grabbed from the about page
# and if data is duplicated that means that the same video was scraped twice
def clean_data(df):
    df = df.dropna(subset=['id'])
    # Drop duplicate rows in column: 'id'
    df = df.drop_duplicates(subset=['id'])
    return df

# Loaded variable 'df' from URI: /Users/gabrielaterrazas/Desktop/erdos-uncreatives-1/raw_data_read_only/data_cleaning/clean_data.csv
#df = pd.read_csv(r'/Users/gabrielaterrazas/Desktop/erdos-uncreatives-1/raw_data_read_only/data_cleaning/clean_data.csv')

df = concatenated_df
df_clean = clean_data(df.copy())

#Save the new cleaned_data CSV file
#output_file = os.path.join(folder_path, "clean_data2.csv")
df_clean.to_csv("../data/clean_data2.csv")
```
