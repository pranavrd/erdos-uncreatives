# Explanation of Data Set

Our data set was obtained by running the Youtube Shorts Data from Apify. We scraped starting on 3/1 over the course of several days. We scraped all posts from at least a year back from a list of influncers. We chose as many influencers as we could with more than 1,000 subscribers who focused on makeup and skincare and who were primarily concerned with product reviews. Our sample is not a perfect random sample (which would be impossible to obtain without access to a full list of youtubers in the beauty niche) but should include the majority of all major influencers from this niche. We ended up with 155 names.

One thing to note about the date data. The time zone was not included when we pulled the data set. To figure out the time zone for the data we posted a test short from a test channel at 9am EST. We then pulled the shorts from that channel and saw that the scraper listed the time of posting as 9am EST. This confirms that the date data in our data set is in EST. Note that everyone who used the scraper to pull the data was also located in the same time zone so local differences did not affect the data.
